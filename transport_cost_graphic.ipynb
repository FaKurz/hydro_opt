```python
import matplotlib.pyplot as plt
```

```python
x_values = range(1,15000)
```

```python
pipeline_rep_min = [0.00008*x*1.1 for x in x_values]
pipeline_rep_max = [0.00025*x*1.1 for x in x_values]
pipeline_new_min = [0.00016*x*1.1 for x in x_values]
pipeline_new_max = [0.0007*x*1.1 for x in x_values]
```

```python
ammonia_min = [2.25 + 0.000025*x for x in x_values]
ammonia_max = [4 + 0.00005*x for x in x_values]

lohc_min = [1.6 + 0.00009*x for x in x_values]
lohc_max = [2.75 + 0.000115*x for x in x_values]

lh2_min = [2.2 + 0.00018*x for x in x_values]
lh2_max = [4.7 + 0.00028*x for x in x_values]
```

SMALL_SIZE = 8
MEDIUM_SIZE = 10
BIGGER_SIZE = 12

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

```python
fig, ax = plt.subplots(2,1, figsize=(12,10))
fig.suptitle("Transportkosten von Wasserstoff nach Transportart")

ax[0].set_title("minimal")
ax[0].set_xlabel("km")
ax[0].set_ylabel("USD / kgH2")

ax[0].plot(x_values, pipeline_rep_min, 'b', label='Pipeline umgebaut')
ax[0].plot(x_values, pipeline_new_min, 'c', label='Pipeline neu')
ax[0].plot(x_values, ammonia_min, 'g', label='Ammoniak')
ax[0].plot(x_values, lohc_min, 'r', label='LOHC')
ax[0].plot(x_values, lh2_min, 'm', label='LH2')

ax[1].set_title("maximal")
ax[1].set_xlabel("km")
ax[1].set_ylabel("USD / kgH2")

ax[1].plot(x_values, pipeline_rep_max, 'b--', label='Pipeline umgebaut')
ax[1].plot(x_values, pipeline_new_max, 'c--', label='Pipeline neu')
ax[1].plot(x_values, ammonia_max, 'g--', label='Ammoniak')
ax[1].plot(x_values, lohc_max, 'r--', label='LOHC')
ax[1].plot(x_values, lh2_max, 'm--', label='LH2')

ax[0].legend()
ax[1].legend()

plt.tight_layout()
plt.savefig('transport costs')
```
